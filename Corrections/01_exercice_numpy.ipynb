```python
# Exercices numériques avec NumPy

import numpy as np
import matplotlib.pyplot as plt

# Définition de la fonction
def f(x):
    return x**4 -2*x**2 + 3

# On coupe l'intervalle pour éviter la discontinuité en x = 0
x = np.linspace(-2, 2, 1000)
y = f(x)
#fprime = np.gradient(y, x)

plt.figure(figsize=(9, 3))

# Courbe de f(x)
plt.plot(x, y, label="f")
#plt.plot(x, fprime, label="f'")

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Représentation de f(x)=x^4 -2x^2+3")
plt.ylim(1.5, 5)  # visualisation plus claire
plt.grid(True)
plt.legend()

plt.show()
```

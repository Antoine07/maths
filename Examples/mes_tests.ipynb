```python
import numpy as np
import matplotlib.pyplot as plt

# Define function and derivatives
x = np.linspace(-3, 3, 400)
f = x**4 - 3*x**2 + 2
f1 = 4*x**3 - 6*x
f2 = 12*x**2 - 6

# Plot
plt.figure(figsize=(8,5))
plt.plot(x, f, label="f(x)")
plt.plot(x, f1, label="f'(x)")
plt.plot(x, f2, label="f''(x)")
plt.axhline(0)
plt.legend()
plt.xlabel("x")
plt.ylabel("value")
plt.title("Function, First Derivative, Second Derivative")
plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------------------------------------
# Dfinition de la fonction f(x,y) = x^2 + y^2
# -------------------------------------------------------------

x = np.linspace(-2, 2, 30)
y = np.linspace(-2, 2, 30)

# Grille de points
X, Y = np.meshgrid(x, y)

# Fonction
Z = X**2 + Y**2

# Gradient de f : f = (2x, 2y)
Fx = 2 * X
Fy = 2 * Y

# -------------------------------------------------------------
# Trac des courbes de niveau + champ de gradient
# -------------------------------------------------------------

plt.figure(figsize=(8,8))

# Courbes de niveau (niveau d'altitude de f)
plt.contour(X, Y, Z, levels=15, cmap="coolwarm")

# Champ de vecteurs du gradient
plt.quiver(X, Y, Fx, Fy, color="black")

# Exemple au point (1,1)
px, py = 1, 1
gx, gy = 2*px, 2*py
plt.scatter(px, py, color="red", s=60, label="Point (1,1)")

# Flche du gradient au point (1,1)
plt.arrow(px, py, gx*0.2, gy*0.2, color="red", width=0.01)

plt.title("Gradient de f(x,y) = x^2 + y^2 :\nDirection d'augmentation la plus rapide")
plt.xlabel("x")
plt.ylabel("y")
plt.axis("equal")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# -------------------------------------------------------------
# Dfinition de la fonction f(x,y) = x^2 + y^2
# -------------------------------------------------------------

# Grille pour reprsenter la surface
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# Point o calculer le gradient
px, py = 1, 1
pz = px**2 + py**2  # valeur de f(1,1)

# Gradient en (1,1)
gx = 2 * px
gy = 2 * py
gz = gx + gy  # composante verticale pour visualisation (optionnelle)

# -------------------------------------------------------------
# Trac de la surface 3D
# -------------------------------------------------------------

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Surface 3D
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)

# Point (1,1,f(1,1))
ax.scatter(px, py, pz, color="red", s=60, label="Point (1,1)")

# Flche reprsentant le gradient en 3D
# On ajoute une petite composante verticale pour quelle soit visible
ax.quiver(px, py, pz, gx, gy, 0.5, color='red', linewidth=2)

# Labels
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x,y)")

ax.set_title("Gradient en 3D pour f(x,y) = x^2 + y^2")

ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # pour activation du mode 3D

# -------------------------------------------------------------
# Dfinition de la fonction
# -------------------------------------------------------------
def f(x, y):
    return x**2 + y**2 - 4*x

# -------------------------------------------------------------
# Cration d'une grille (x,y)
# -------------------------------------------------------------
x = np.linspace(-1, 5, 300)   # intervalle large pour visualiser la convexit
y = np.linspace(-3, 3, 300)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# -------------------------------------------------------------
# Point critique thorique
# -------------------------------------------------------------
xc, yc = 2, 0
zc = f(xc, yc)

# -------------------------------------------------------------
# FIGURE 1 : Surface 3D
# -------------------------------------------------------------
fig = plt.figure(figsize=(12, 5))
ax = fig.add_subplot(121, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.9)
ax.scatter(xc, yc, zc, color="red", s=60, label="Minimum (2,0)")
ax.set_title("Surface 3D de f(x,y) = x + y - 4x")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x,y)")
ax.legend()

# -------------------------------------------------------------
# FIGURE 2 : Courbes de niveau (Contours)
# -------------------------------------------------------------
ax2 = fig.add_subplot(122)
contours = ax2.contour(X, Y, Z, levels=25, cmap='viridis')
ax2.clabel(contours)
ax2.scatter(xc, yc, color="red", s=50)
ax2.set_title("Courbes de niveau (minimum marqu)")
ax2.set_xlabel("x")
ax2.set_ylabel("y")

plt.tight_layout()
plt.show()
```

```python
import sympy as sp
import numpy as np

# Dfinition des variables
x, y, z = sp.symbols('x y z')

# Dfinition de la fonction
f = x**3 - 3*x + y**2 + x*y - z**2 + 4*z

# Gradient
grad_f = [sp.diff(f, var) for var in (x,y,z)]

print("Gradient :")
print(grad_f)

# Rsolution du systme grad f = 0
crit_points = sp.solve(grad_f, (x,y,z), dict=True)
print("\nPoints critiques :")
print(crit_points)

# Hessienne
H = sp.hessian(f, (x,y,z))
print("\nHessienne symbolique :")
sp.pprint(H)

# Analyse numrique des valeurs propres
print("\nClassification avec valeurs propres :")
for p in crit_points:
    H_num = np.array(H.subs(p), dtype=float)
    eigs = np.linalg.eigvals(H_num)
    print("\nPoint :", p)
    print("Hessienne numrique :\n", H_num)
    print("Valeurs propres :", eigs)
    
    if all(eigs > 0):
        print("Conclusion : MINIMUM local.")
    elif all(eigs < 0):
        print("Conclusion : MAXIMUM local.")
    elif any(e == 0 for e in eigs):
        print("Conclusion : test inconcluant.")
    else:
        print("Conclusion : POINT SELLE.")
```
